```python
def remove_english_words(lines):
    sentences = []
    for i in range(len(lines)):
        temp = lines[i]
        lst = temp.split()
        flag = 0
        for word in lst:
            x = re.search(r"^[a-zA-Z0-9]", word)
            if x:
                flag += 1
        if flag == 0:
            sentences.append(temp)
    return sentences
```
