```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Dataset 3.csv')
df['tweets']
```

```python
df.dtypes
```

```python
all_tweets = []
for item in df['tweets']:
    all_tweets.append(item)
```

```python
all_tweet_sentiment = []
for item in df['sentiment']:
    all_tweet_sentiment.append(item)
```

```python
def remove_punctuation(lines):
    punctuations = '''!()-[]।{};:'"\,<>./?@#$%^&*_~'''
    punch_remove_line = []
    # remove punctuation from the string
    for i in range(len(lines)):
        no_punct = ""
        for char in lines[i]:
            if char not in punctuations:
                no_punct = no_punct + char
        punch_remove_line.append(no_punct)
    return punch_remove_line
```

```python
punc_tweet = remove_punctuation(all_tweets)
```

```python
punc_tweet
```

```python
len(all_tweet_sentiment)
```

```python
punc_df = pd.DataFrame({'tweets' : punc_tweet, 'sentiment' : all_tweet_sentiment})
punc_df
```

```python
punc_df.to_csv('without_punctuation.csv')
```

```python
without_punc = pd.read_csv('without_punctuation.csv', index_col = False)
without_punc.drop(['Unnamed: 0'], inplace = True, axis = 1)
without_punc
```

```python
without_punc['tweets'] = without_punc['tweets'].str.replace("@", "")
without_punc['tweets'] = without_punc['tweets'].str.replace(r"http\S+", "")
without_punc['tweets'] = without_punc['tweets'].str.replace("[^a-zA-Z]", " ")
without_punc['tweets'] = without_punc['tweets'].str.replace("  ", " ")
without_punc['tweets'] = without_punc['tweets'].str.strip()
```

```python
without_punc
```

```python
without_punc.to_csv('clean_tweets.csv')
```
